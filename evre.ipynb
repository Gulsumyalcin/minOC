```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# CSV dosyasƒ±nƒ± oku
veri = pd.read_csv(("./S17_not.csv"))
# time s√ºtununu i√ßeren bir DataFrame olu≈ütur
time_df = veri[['time']]
flux_degerleri = veri["flux"]



# T0 hesapla
T0 = 1765.836456
print(f"T0: {T0}")

d√∂nem =  7.557813

print(f"d√∂nem: {d√∂nem}")

evre= (time_df - T0)/d√∂nem

evre2 = evre - np.floor(evre)

print(f"evre: {evre2}")

# Nokta daƒüƒ±lƒ±m grafiƒüi √ßiz
plt.scatter(evre2, flux_degerleri)
plt.xlabel('Evre')
plt.ylabel('Flux')
plt.title('Evre vs Flux Daƒüƒ±lƒ±m Grafiƒüi')
plt.show()
```
